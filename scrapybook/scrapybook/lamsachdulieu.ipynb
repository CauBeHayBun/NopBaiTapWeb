```python
from pyspark.sql import SparkSession

# Tạo Spark session
spark = SparkSession.builder \
    .appName("Data Processing") \
    .config("spark.mongodb.input.uri", "mongodb://localhost:27017/myscrapybook_db.books") \
    .config("spark.mongodb.output.uri", "mongodb://localhost:27017/myscrapybook_db.books") \
    .getOrCreate()

# Đọc dữ liệu từ MongoDB
df = spark.read.format("mongod").load()
```

```python
import pandas as pd
from pymongo import MongoClient
from sqlalchemy import create_engine

# Kết nối đến MongoDB
client = MongoClient('mongodb://localhost:27017/')
# client = MongoClient('mongodb://mongodbbook:27017/')
db = client['myscrapybook_db']
collection = db['books']

# Lấy dữ liệu từ MongoDB
data = list(collection.find({}))
df = pd.DataFrame(data)
```

```python
# Hiển thị dữ liệu để kiểm tra
print(df.head())  # Hiển thị 5 dòng đầu tiên

# Kiểm tra kiểu dữ liệu của các cột
print(df.dtypes)

```

```python
df['noidung_brand'] = df['noidung_brand'].str.replace('Mua sách online tại Bookbuy.vn và nhận nhiều ưu đãi.', '', regex=False)
```

```python
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
#này là xóa khoảng trắng đầu cuối all cột
```

```python
df['giahientai'] = df['giahientai'].str.replace(r'[^\d]', '', regex=True)
```

```python
df['giathitruong'] = df['giathitruong'].str.replace(r'[^\d]', '', regex=True)
```

```python
df['sotrang'] = df['sotrang'].str.replace(r'[^\d]', '', regex=True)
```

```python
df['trongluong'] = df['trongluong'].str.replace(r'[^\d]', '', regex=True)
```

```python
from sqlalchemy import create_engine

# Tạo engine kết nối đến PostgreSQL
DATABASE_URL = 'postgresql://localhost:5432/mydatabase'  # Không cần username và password
engine = create_engine(DATABASE_URL)

# Kiểm tra kết nối
with engine.connect() as connection:
    result = connection.execute("SELECT version();")
    for row in result:
        print(row)
```

```python
# Giả sử DataFrame df có các trường tương ứng với bảng trong PostgreSQL
df.to_sql('books', engine, if_exists='append', index=False)
```
